```python
def checkX(x,y,l,h,dctX,dctY):
    
    retVal =[]
    if dctX[(x,y,h)] == 1:
        return retVal
    else:
        dctX[(x,y,h)] = 1
    
    if x+h <0:
        return retVal
    if x+h >= len(l[y]):
        return retVal
    
    if  x+h >= len(l[y]) or x+h <0:
        return retVal

    while l[y][x] < l[y][x+h]:
        retVal.append((y,x))
        retVal = retVal + checkY(x,y,l,-1,dctX,dctY)
        retVal = retVal + checkY(x,y,l,1,dctX,dctY)
        x = x +h
        if x == len(l[y])-1 or x ==0:
            if l[y][x-h] < l[y][x]:
                retVal.append((y,x-h))
            return retVal
    return retVal
    
def checkY(x,y,l,h,dctX,dctY):
    retVal =[]
    if dctY[(x,y,h)] == 1:
        return retVal
    else:
        dctY[(x,y,h)] = 1
    if y+h <0:
        return retVal
    if y+h >= len(l):
        return retVal
    if  y+h >= len(l) or y+h <0:
        return retVal
    while l[y][x] < l[y+h][x]:
        retVal.append((y,x))
        retVal = retVal + checkX(x,y,l,-1,dctX,dctY)
        retVal = retVal + checkX(x,y,l,1,dctX,dctY)
        y = y +h
        if y == len(l)-1 or y ==0:
            if l[y-h][x] < l[y][x]:
                retVal.append((y-h,x))
            return retVal
    return retVal  

        
            
```

```python
def checkX2(x,y,l,h,dctX,dctY):
    retVal = []
    if dctX[(x,y,h)] == 1:
        return retVal
    else:
        dctX[(x,y,h)] = 1
    if x< 0:
        return retVal
    while l[y][x] < 9:
        retVal.append((x,y))
        x = x+h
        if x >= len(l[y]) or x <0:
            return retVal
        retVal = retVal +checkY2(x,y,l,1,dctX,dctY)
        retVal = retVal + checkY2(x,y,l,-1,dctX,dctY)
    return retVal
def checkY2(x,y,l,h,dctX,dctY):
    retVal = []
    if dctY[(x,y,h)] == 1:
        return retVal
    else:
        dctY[(x,y,h)] = 1
    if y< 0:
        return retVal
    while l[y][x] < 9:
        retVal.append((x,y))
        y = y+h
        if y >= len(l) or y <0:
            return retVal
        retVal = retVal +checkX2(x,y,l,1,dctX,dctY)
        retVal = retVal + checkX2(x,y,l,-1,dctX,dctY)    
    return retVal
```

```python
from collections import defaultdict
f = open("input14-1.txt", "r")
l = f.read().split("\n")[:-1]
ll = [ x  for x in l]
t =[]
print(ll)
#print(t)
pattern = ll[0]
rules = list(set([x for x in ll[2:]]))
print(rules)
```

```python
def check(s, p):
    tmp = [x for x in p if s + ' -> ' in x]
    if tmp != []:
        return tmp[0].split(' -> ')[1]
    return ''
```

```python
print(check("NN",rules))
```

```python
newPattern = ""
for n in range(0,10):
    if newPattern !="":
        pattern = newPattern
    newPattern = ""
    for i in range(0,len(pattern)-1):
        tmp = check(pattern[i:i+2],rules)
        #print(pattern[i:i+2],pattern[i] + tmp +pattern[i+1])
        if newPattern == '':
            newPattern += pattern[i] + tmp +pattern[i+1]
        else:
            newPattern += tmp +pattern[i+1]
print(newPattern)
ctList=[]
for x in set(newPattern):
    ctList.append(newPattern.count(x))
print(max(ctList)- min(ctList))
```

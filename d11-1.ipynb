```python
def incAdj(x,y,l,dct):
    retVal =[]
    for _x in range(x-1,x+2):

        if _x < 0 or _x >= len(l):
            continue
        for _y in range(y-1,y+2):
            if _y < 0 or _y >= len(l):
                continue
            if _x == x and _y == y:
                continue
            if l[y][x] >9:
                
                if l[_y][_x] <= 9 and dct[(y,x)] == 0 and dct[(_y,_x)] == 0:
                    l[_y][_x] +=1
                    #print((y,x),(_y,_x),l[_y][_x])
            #retVal.append((_x,_y))
    return retVal
```

```python
def inc(l):
    for i in range(0,len(l)):
        l[i]+=1
```

```python
def res(l,dct):
    ct = 0
    for y in range(0,len(l)):
        for x in range(0,len(l[y])):
            if l[y][x] >9:
                
                incAdj(x,y,l,dct)
                dct[(y,x)] = 1
                l[y][x] = 0
                ct+=1
    return ct
```

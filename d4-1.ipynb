```python
def check(l):
    for x in l:
        if x.count(-1) == len(x):
            return True
    for i in range(0,len(l[0])):
        match = True
        for x in range(0,len(l)):
         
            if l[x][i] != -1:
                match = False
                break
        if match:
            return True
            
        
    return False
```
